# **Introduction**

### Data card :

This Data set was collected by a survey conducted by Google forms from University student in order to examine their current academic situation and mental health.

#### Attributes :

1. **Timestamp:** The date and time when the data entry was recorded in the format "Month/Day/Year Hour:Minute."
2. **Choose your gender:** The gender of the individual, specified as "Female" or "Male."
3. **Age:** The age of the individual.
4. **What is your course?:** The field of study or course that the individual is pursuing, such as "Engineering," "Islamic education," "BIT" , "Laws," and "Mathematics."
5. **Your current year of Study:** The academic year the individual is currently in, such as "year 1," "year 2," "year 3," and "year 4."
6. **What is your CGPA?:** The Cumulative Grade Point Average of the individual, specified as a range (e.g., "3.00 - 3.49").
7. **Marital status:** The marital status of the individual, indicated as "Yes" or "No."
8. **Do you have Depression?:** Whether the individual has depression, indicated as "Yes" or "No."
9. **Do you have Anxiety?:** Whether the individual has anxiety, indicated as "Yes" or "No."
10. **Do you have Panic attack?:** Whether the individual experiences panic attacks, indicated as "Yes" or "No."
11. **Did you seek any specialist for treatment?:** Whether the individual sought any specialist for treatment, indicated as "Yes" or "No."

Each row in your dataset represents a unique individual's responses to these attributes. The dataset captures information about students, including their demographics, academic details, and mental health-related information.

# **Data Cleaning**

```python
df.dropna(inplace=True)
df.isna().sum()
```

*we dropped the whole raw containing a missing value*

```python
columns_names = ['Timestamp', 'Gender', 'Age', 'Course', 'Year of Study', 'CGPA', 'Is Married?', 'Has Depression?','Has Anxiety?', 'Has Panic Attacks?', 'Is Undergoing Treatment?']
df.columns = columns_names
df.info()
```

*The naming of each columns was too long so it's just easier to rename to shorter and more descriptive names.*

### **Age**

```python
# dtype('float64') to in
df['Age'] = df['Age'].astype(int)
df['Age'].dtype
```

### **Year of Study**

```python
df['Year of Study'].unique()
# problem in the case !!!!!!! redandonce
```

```python
#solution
df['Year of Study'] = df['Year of Study'].str.split().str[1].astype(int)
df['Year of Study'].unique()
```

### **CGPA Column**

```python
df['CGPA'].unique()
```

```python
#Getting rid of different values cause by white spaces
df['CGPA'] = df['CGPA'].str.strip()
df['CGPA'].unique()
```

### **Students Conditions**

```python
depressed_filter = df['Has Depression?'] == 'Yes'
anxiety_filter = df['Has Anxiety?'] == 'Yes'
panicking_filter = df['Has Panic Attacks?'] == 'Yes'
conditions_filter = anxiety_filter | depressed_filter | panicking_filter
no_conditions_filter = ~depressed_filter & ~anxiety_filter & ~panicking_filter
```

```python
#After creating some filters, I added a new column called Has Condition? to indicate wheter a Student is affected or not by a condition.
df['Has Condition?'] = np.where(conditions_filter, 'Yes',"No")
df
```

For each condition (**Has Depression**, **Has Anxiety** or **Has Panic Attacks**) a student has, a letter (**A** = Anxiety, **D** = Depression and **P** = Panic Attacks) gets appended to the Condition column making it easier to keep track and filter data.

```python
df['Condition'] = ''
df.loc[anxiety_filter, 'Condition'] += 'A'
df.loc[depressed_filter, 'Condition'] += 'D'
df.loc[panicking_filter, 'Condition'] += 'P'
df.loc[no_conditions_filter, 'Condition'] += 'None'
# df['Condition'].unique()

df
```

# **Initial Data Visualization** and **Analysis**

### **Students by Gender**

```python
labels = df['Gender'].value_counts().index

plt.pie(df['Gender'].value_counts(),labels=labels, autopct='%1.1f%%',startangle=180, explode=(0.005,0.005))
plt.axis('equal')
plt.title('Students per Gender', fontsize=12, fontweight='bold', pad=20)
plt.show()
```

### **Students by Year of Study**

```python
sns.countplot(data=df, x='Year of Study').set(xlabel=None, ylabel='Stuents')
plt.title('Students per Year of Study', fontsize=12, fontweight='bold', pad=20)
sns.color_palette("pastel")

plt.show()
```

### **Students by Age**

```python
sns.countplot(data=df, x='Age')
plt.xlabel('Age')
plt.ylabel('Students')
plt.title('Age Distribution', fontsize=12, fontweight='bold', pad=20)

plt.show()
```
